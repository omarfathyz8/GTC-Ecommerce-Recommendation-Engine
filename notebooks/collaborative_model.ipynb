# Feature engineering

```python
df["helpfulness_ratio"] = df.apply(
    lambda x: x["HelpfulnessNumerator"]/x["HelpfulnessDenominator"]
    if x["HelpfulnessDenominator"] > 0 else 0, axis=1)
df["score_weight"]=df["helpfulness_ratio"]*df["Score"]
df["score_weight"].describe()
```

```python
year=df["year"].max()
df["time_weight"]=1/(1+(year-df["year"]))
df["score_weight2"]=df["time_weight"]*df["Score"]
df["score_weight2"].describe()
```

```python
df["fianl_score"]=df["helpfulness_ratio"]*df["time_weight"]*df["Score"]
df["final_score2"]=np.log1p(df["helpfulness_ratio"])*df["time_weight"]*df["Score"]
df["final_score3"] = np.sqrt(df["helpfulness_ratio"])*df["time_weight"]*df["Score"]
```

# Modeling

```python
from surprise import Dataset, Reader
data_cf=df[["UserId","ProductId","final_score2"]]
reader=Reader(rating_scale=(1,5))  
dataset=Dataset.load_from_df(data_cf,reader)
```

```python
from surprise.model_selection import train_test_split
trainset,testset =train_test_split(dataset,test_size=0.2,random_state=42)
```

```python
from surprise import SVD
from surprise import accuracy
model=SVD()
model.fit(trainset)
predictions=model.test(testset)
accuracy.rmse(predictions)
accuracy.mae(predictions)
```

```python
from surprise.model_selection import RandomizedSearchCV
param_dist = {
    'n_factors':[50, 100, 150, 200,250],
    'n_epochs':[20, 40, 60, 80],
    'lr_all':np.linspace(0.002, 0.02,5),
    'reg_all':np.linspace(0.02, 0.1,5)}
rs=RandomizedSearchCV(SVD, param_distributions=param_dist,measures=['rmse','mae'],cv=3,n_iter=10,random_state=42)
rs.fit(dataset)
print("Best RMSE score:", rs.best_score['rmse'])
print("Best hyperparameters:", rs.best_params['rmse'])
```

```python
from surprise import SVD
from surprise import accuracy
model=SVD(n_factors=50,n_epochs=80,lr_all= 0.02,reg_all= 0.08)
model.fit(trainset)
predictions=model.test(testset)
accuracy.rmse(predictions)
accuracy.mae(predictions)
```
